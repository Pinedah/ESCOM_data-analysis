```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
iris.data
```

```python
for i in range(0, 4):
    print(f"{iris.feature_names[i]} median: {np.median(iris.data[:, i])}")
```

```python
petal_widths = iris.data[:, 3] # se obtiene el arrglo de valores para el ancho del petalo

valores_intervalo = np.arange(0.1, 2.5, 0.1) # se define el intervalo de 0.1 en 0.1 desde 0 hasta 2.5
array_relativa = []
array_acumulada = []

acumulada = 0
for valor_intervalo in valores_intervalo:
    relativa = 0
    for petal_width in petal_widths:
        if valor_intervalo == petal_width:
            relativa = relativa + 1
    array_relativa.append(relativa)
    acumulada = acumulada + relativa
    array_acumulada.append(acumulada)

# Crear DataFrame
tabla = pd.DataFrame({
   'Valor': valores_intervalo,
   'Total': array_relativa,
   'Acumulado': array_acumulada
})

tabla
```
