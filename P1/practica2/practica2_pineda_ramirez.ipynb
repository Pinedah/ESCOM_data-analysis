Remover los outliers de la variable HR y comparar la correlación entre las variables R, HR, SO e IBB antes y después removerlos.

```python
df_outliers = df_parte1.copy()
```

```python
df_outliers['HR'].quantile(0.75)
```

```python
def calcular_outliers(df):
    Q1 = df['HR'].quantile(0.25)
    Q3 = df['HR'].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5*IQR
    limite_inferior = Q1 - 1.5*IQR
    outliers = ((df['HR'] < limite_inferior) |( df['HR'] > limite_superior))
    outliersnt = df[~outliers]
    return outliersnt    
```

```python
df_outliers = calcular_outliers(df_parte1)
```

```python
subset1 = df_parte1[['R', 'HR', 'SO', 'IBB']]
corr1 = subset1.corr()
subset2 = df_outliers[['R', 'HR', 'SO', 'IBB']]
corr2 = subset2.corr()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(corr1, annot=True, cmap='coolwarm', center=0, fmt=".2f", ax=axes[0])
axes[0].set_title("Con outliers")
sns.heatmap(corr2, annot=True, cmap='coolwarm', center=0, fmt=".2f", ax=axes[1])
axes[1].set_title("Sin outliers")

plt.tight_layout()
plt.show()
```

Explicar por qué cambia tanto la correlación entre HR e IBB después de
 remover los outliers.

La correlación en el dataframe original es mayor porque al dar pase a la primera base de manera intencional aumenta la posibilidad de hacer más home runs. En cambio, al quitar los outliers en la variable asociada a los Home Runs, disminuye la correlación entre ambos ya que disminuye la posibilidad de hacer un Home Run.

```python
df_parte2 = df.copy()
```

```python
df_parte2['yearID'].unique()
```

```python
promedios = df_parte2.groupby('yearID')['HR'].mean()
print(promedios)
```

```python
series = pd.DataFrame({
    
    'avgHR': promedios
})
series
```

```python
series = series.reset_index()
series
```

```python
def filtro_mediana(df, columna, q):    
    valores = df[columna].values
    n = len(valores)
    resultado = []

    for i in range(n):
        # ventana de i-q hasta i (asimétrica hacia atrás)
        inicio = max(0, i - q)
        fin = i + 1
        w_kq = valores[inicio:fin]
        
        # mediana de la ventana
        m_kq = np.median(w_kq)
        resultado.append(m_kq)
    
    df_filtrado = df.copy()
    df_filtrado[f'mediana_asim_q{q}'] = resultado
    return df_filtrado
```

```python
medianas = filtro_mediana(series, 'avgHR', 20)
```

```python
medianas.head()
```

```python
def filtro_exponencial(df, columna, theta):    
    valores = df[columna].values
    n = len(valores)
    y = [valores[0]]  # inicializamos con el primer valor
    
    for k in range(1, n):
        y_k = y[k-1] + theta * (valores[k] - y[k-1])
        y.append(y_k)
    
    df_filtrado = df.copy()
    df_filtrado[f'exp_smooth_theta{theta}'] = y
    return df_filtrado
```

```python
exponencial = filtro_exponencial(series, 'avgHR', 0.1)
```

```python
exponencial.head()
```
