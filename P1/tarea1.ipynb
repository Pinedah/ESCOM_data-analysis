Con lo anterior podemos identificar que la variable la cual posee una media de 1.3 es la relacionada a "petal width (cm)", la cuál se encuentra en la posición 3 del conjunto de datos cargado.

Sabiendo lo anterior podemos proceder a generar la función la cuál será usada para el analisis en la primer tabla solicitada.

```python
def generar_tabla(caracteristica_iris):
    # PRIMER PARTE DE LA TABLA
    valores_intervalo1 = np.arange(min(caracteristica_iris), round(np.median(caracteristica_iris + 0.1), 1), 0.1) # se define el intervalo de 0.1 en 0.1 desde 0 hasta 2.5
    valores_intervalo1 = [round(valor, 1) for valor in valores_intervalo1]

    array_relativa1 = []
    array_acumulada1 = []

    acumulada1 = 0
    for valor_intervalo in valores_intervalo1:
        relativa = 0
        for petal_width in sorted(caracteristica_iris)[:75]:
            if valor_intervalo == petal_width:
                relativa = relativa + 1
        array_relativa1.append(relativa)
        acumulada1 = acumulada1 + relativa
        array_acumulada1.append(acumulada1)

    # SEGUNDA PARTE DE LA TABLA
    valores_intervalo2 = np.arange(max(caracteristica_iris), round(np.median(caracteristica_iris - 0.1), 1), -0.1) # se define el intervalo de 0.1 en 0.1 desde 0 hasta 2.5
    valores_intervalo2 = [round(valor, 1) for valor in valores_intervalo2]

    array_relativa2 = []
    array_acumulada2 = []

    acumulada2 = 0
    for valor_intervalo in valores_intervalo2:
        relativa = 0
        for petal_width in sorted(caracteristica_iris)[75:]:
            if valor_intervalo == petal_width:
                relativa = relativa + 1
        array_relativa2.append(relativa)
        acumulada2 = acumulada2 + relativa
        array_acumulada2.append(acumulada2)

    # Crear DataFrame
    tabla = pd.DataFrame({
    'Valor': valores_intervalo1,
    'Total': array_relativa1,
    'Acumulado': array_acumulada1,
    'Valor2': valores_intervalo2,
    'Total2': array_relativa2,
    'Acumulado2': array_acumulada2
    })

    return tabla
```

Obtenemos el vector del ancho de los petalos (petal_widths) usnado slicing.

## Tabla 1 (ancho de los petalos): 

```python
petal_widths = iris.data[:, 3] # se obtiene el arrglo de valores para el ancho del petalo
generar_tabla(petal_widths)
```

Creamos una segunda función para la tabla dos ya que las indicaciones especificas de la tabla 1 y las diferencias en la distribución de los datos a partir de las medianas no coincide al generar los tamaños iguales de los arreglos.

```python
def generar_tabla2(caracteristica_iris):
    # PRIMER PARTE DE LA TABLA
    valores_intervalo1 = np.arange(min(caracteristica_iris), (min(caracteristica_iris) + max(caracteristica_iris))/2, 0.1) # se define el intervalo de 0.1 en 0.1 desde 0 hasta 2.5
    valores_intervalo1 = [round(valor, 1) for valor in valores_intervalo1]

    array_relativa1 = []
    array_acumulada1 = []

    acumulada1 = 0
    for valor_intervalo in valores_intervalo1:
        relativa = 0
        for petal_width in sorted(caracteristica_iris)[:75]:
            if valor_intervalo == petal_width:
                relativa = relativa + 1
        array_relativa1.append(relativa)
        acumulada1 = acumulada1 + relativa
        array_acumulada1.append(acumulada1)

    # SEGUNDA PARTE DE LA TABLA
    valores_intervalo2 = np.arange(max(caracteristica_iris), (min(caracteristica_iris) + max(caracteristica_iris))/2, -0.1) # se define el intervalo de 0.1 en 0.1 desde 0 hasta 2.5
    valores_intervalo2 = [round(valor, 1) for valor in valores_intervalo2]

    array_relativa2 = []
    array_acumulada2 = []

    acumulada2 = 0
    for valor_intervalo in valores_intervalo2:
        relativa = 0
        for petal_width in sorted(caracteristica_iris)[75:]:
            if valor_intervalo == petal_width:
                relativa = relativa + 1
        array_relativa2.append(relativa)
        acumulada2 = acumulada2 + relativa
        array_acumulada2.append(acumulada2)

    # Crear DataFrame
    tabla = pd.DataFrame({
    'Valor': valores_intervalo1,
    'Total': array_relativa1,
    'Acumulado': array_acumulada1,
    'Valor2': valores_intervalo2,
    'Total2': array_relativa2,
    'Acumulado2': array_acumulada2
    })

    return tabla
```
